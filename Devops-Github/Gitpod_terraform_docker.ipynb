 Set your repo path:

```python
repo_path = "/content/drive/MyDrive/Devops/Devops-Github"
```

Now create .gitpod.yml and .gitpod.Dockerfile:

```python
import os

os.makedirs(repo_path, exist_ok=True)

# .gitpod.yml

with open(os.path.join(repo_path, ".gitpod.yml"), "w") as f:
  f.write('''image:
  file: .gitpod.Dockerfile

tasks:
  - name: Setup Workspace
    init: |
      echo "✅ Workspace ready for Terraform + Docker"
''')

# .gitpod.Dockerfile

with open(os.path.join(repo_path, ".gitpod.Dockerfile"), "w") as f:
  f.write('''FROM gitpod/workspace-full

# Install Docker and Terraform
RUN apt-get update && \
    apt-get install -y docker.io unzip curl && \
    curl -fsSL https://apt.releases.hashicorp.com/gpg | gpg --dearmor -o /usr/share/keyrings/hashicorp-archive-keyring.gpg && \
    echo "deb [signed-by=/usr/share/keyrings/hashicorp-archive-keyring.gpg] https://apt.releases.hashicorp.com $(lsb_release -cs) main" | tee /etc/apt/sources.list.d/hashicorp.list && \
    apt update && apt install terraform -y
''')

print("✅ Gitpod config files created successfully.")
```
