```python
shipment_data = {
    "shipment": {
        "id": "SHIP_1234",
        "carrier": {
            "id": "CARR_56",
            "mode": "TL"
        },
        "cost": 452.75,
        "status": "Scheduled"
    },
    "tracking": {
        "events": [
            {"location": "DC01", "time": "2024-01-01T10:00", "status": "In Transit"},
            {"location": "WH02", "time": "2024-01-02T14:00", "status": "Delayed"},
            {"location": "Store07", "time": "2024-01-03T18:00", "status": "Delivered"}
        ]
    }
}

# Task 1 - Print shipment ID + carrier ID
print(shipment_data['shipment']['id'] + " " + shipment_data['shipment']['carrier']['id'])

# Task 2 - Update status to "Delivered"
shipment_data['shipment']['status'] = "Delivered"
print(shipment_data['shipment']['status'])

# Task 3 - Add a new key delivered_at = "2025-04-22T15:30" inside "shipment"
shipment_data['shipment']['delivered_at'] = "2025-04-22T15:30"
print(shipment_data['shipment']['delivered_at'])

# Task 4 - Loop through tracking["events"] and print only non-delivered statuses
for event in range(len(shipment_data['tracking']['events'])) :
  if shipment_data['tracking']['events'][event]['status'] != "Delivered" :
    print(shipment_data['tracking']['events'][event]['status'])

# Task 5 - Count how many events had status "In Transit"
intransit_count = 0
for event in range(len(shipment_data['tracking']['events'])) :
  if shipment_data['tracking']['events'][event]['status'] == "In Transit" :
    intransit_count += 1

print(intransit_count)
```
